### Activity Selection Problem

let us learn Activity Selection Problem using an real life use case
In the single-machine scheduling problem, we are given a set of n activities Ai. Each job i has a starting time si, deadline di and profit pi. At any time instant, we can do only one job. Doing a job i earns a profit pi. Generate a solution to select the
largest set of mutually compatible jobs and calculate the total profit generated by the machine.

```python
def activity_selection(acts,n):
    acts = sorted(acts, key = lambda x:x[2])
    selected = list()
    #first activity always selected
    selected.append(acts[0])
    k = 0
    for i in range(1,n):
        if(acts[i][1] > acts[k][2]):
            selected.append(acts[i])
            k = i
    #profit calculation
    profit = 0
    for i in selected:
        profit += i[3]
    return selected, profit
def inputactivity():
    n = int(input("Enter number of activities: "))
    acts = [[0]*(4) for _ in range(n)]
    for i in range(n):
        acts[i][0] = input(f"Enter activity name {i+1}: ")
        acts[i][1] = int(input(f"Enter Start Time {i+1}: "))
        acts[i][2] = int(input(f"Enter Finish Time {i+1}: "))
        acts[i][3] = int(input(f"Enter Profit {i+1}: "))
    return n, acts
```

```python
user = input("How do you want to run the program? 1: User input 2: Sample data ::")
if user == '1':
    #call function for user input
    n,acts = inputactivity()
else:
    #sample data
    acts = [['A1', 1, 4, 10],['A2', 3, 5, 15],['A3', 0, 6, 14],['A4', 5, 7, 12],['A5', 3, 9, 20],
            ['A6', 5, 9, 30],['A7', 6, 10, 32],['A8', 8, 11, 28],['A9', 8, 12, 30],['A10', 2, 14, 40],['A11', 12, 16, 45]]
    n = 11
    print("\nData: ", acts)
selected, profit = activity_selection(acts,n)
print("\nSelected activities are: ", end="")
#print selected activities
for i in selected:
    print(i[0], end=", ")
#print total profit
print("\nTotal Profit: ",profit)
```

### Maximum Possible equal sum of 3 stacks
Given three stacks of the positive numbers, the task is to find the possible equal
maximum sum of the stacks with the removal of top elements allowed.

```python
#program to calculate maximum possible equal sum for 3 stacks
def maxpossum(s1,s2,s3):
    while sum(s1) != sum(s2) or sum(s1) != sum(s3):
        if max(sum(s1), sum(s2), sum(s3)) == sum(s1):
            s1.pop()
        elif max(sum(s1), sum(s2), sum(s3)) == sum(s2):
            s2.pop()
        else:
            s3.pop()
    return s1,s2,s3,sum(s1)
def inputstacks():
    s1 = list(map(int, input("Enter elemets of stack 1: ").split()))
    s2 = list(map(int, input("Enter elemets of stack 2: ").split()))
    s3 = list(map(int, input("Enter elemets of stack 3: ").split()))
    return s1,s2,s3
```

```python
s1,s2,s3 = inputstacks()
s1,s2,s3,msum = maxpossum(s1,s2,s3)
print("The maximum possible equal sum is: ",msum)
print("The stacks are: ")
print("Stack 1: ",s1)
print("Stack 2: ",s2)
print("Stack 3: ",s3)
```

### Application of above program: Lucky Wheel at a casino

Winner is announced if and only if non empty solution of Maximum Possible equal sum exists

```python
#lucky wheel using maximum possible sum
import random
def luckydraw(reward):
    if reward == 1:
        s1 = random.sample(range(1,101),random.randint(10,100))
        s2 = random.sample(range(1,101),random.randint(10,100))
        s3 = random.sample(range(1,101),random.randint(10,100))
    elif reward == 2:
        s1 = random.sample(range(1,1001),random.randint(10,100))
        s2 = random.sample(range(1,1001),random.randint(10,100))
        s3 = random.sample(range(1,1001),random.randint(10,100))
    else:
        s1 = random.sample(range(1,10001),random.randint(10,100))
        s2 = random.sample(range(1,10001),random.randint(10,100))
        s3 = random.sample(range(1,10001),random.randint(10,100))
    s1,s2,s3,msum = maxpossum(s1,s2,s3)
    if(msum != 0):
        print("CONGRAGULATIONS, YOU WIN!!!", reward)
        return 1
    else:
        print("You loose, try again!")
        return 0
```
